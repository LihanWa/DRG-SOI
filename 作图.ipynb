```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据
x = np.linspace(0, 10, 10)  # 生成0-10的10个点
y = np.sin(x)  # 让y是x的正弦
dy = 0.5 + 0.2 * np.sin(x + 1)  # 演示用的误差范围

# 绘图
plt.plot(x, y, 'bo-', label='Curve')
plt.errorbar(x, y, yerr=dy, fmt='o', color='blue', ecolor='gray', elinewidth=1, capsize=3)

# 添加标签等
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Curve with Vertical Error Bars')
plt.legend()

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt

# 创建一个图和一个坐标轴
fig, ax1 = plt.subplots()

# 示例数据
# 100, 250, 500, 750, 1000, 1250,1500,
x = np.array([1000,2000,4000,6000,8000,10000,12000,14000,16000])
MAE = np.array([2.472,1.879,1.447,1.41,1.383,1.287,1.223,1.19,1.165])
ACCURACY = np.array([0.088,0.160,0.242,0.258,0.2669,0.288,0.300,0.304,0.317])
y_sd = np.array([0.002,0.003, 0.0025,0.003,0.002,0.0025,0.0035,0.003, 0.003])  # 示例的标准差
y_sd*=16
print(y_sd)
plt.errorbar(x, MAE, yerr=y_sd, fmt='o', color='#f5426c', capsize=1, label='MAE',markersize=4)


# 在第一个坐标轴上绘制第一条线
ax1.plot(x, MAE, color='#f5426c')
ax1.set_xlabel('Number of Samples',fontsize=18)
ax1.set_ylabel('MAE', color='#f5426c',fontsize=18)
ax1.tick_params(axis='y', labelcolor='#f5426c',labelsize=18)
# ax1.legend(loc='upper left')
xticks = np.append(1000,x)  # Add 1000 to the x-tick values
ax1.set_xticks(x)
ax1.set_xticklabels(x, rotation=45)
ax1.tick_params(axis='x', labelsize=18)  # Adjusted here
ax1.yaxis.set_ticks(np.arange(1.15,2.6, 0.15))

# 创建一个共享x轴的第二个坐标轴
ax2 = ax1.twinx()
print(len(x))
print(len(MAE))
# 在第二个坐标轴上绘制第二条线
ax2.plot(x, ACCURACY, color='b' )
ax2.set_ylabel('ACCURACY', color='b',fontsize=18)
ax2.tick_params(axis='y', labelcolor='b',labelsize=18)
# ax2.legend(loc='upper right')
ax2.yaxis.set_ticks(np.arange(0.09,0.35, 0.03))

y_sd*=.25
plt.errorbar(x, ACCURACY, yerr=y_sd, fmt='o', color='b', capsize=1, label='ACCURACY',markersize=4)
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left', bbox_to_anchor=(0.05, 1),fontsize=14)
# 显示图表
# plt.title('Plotting Two Lines with Two Y-Axes')
plt.show()
```
